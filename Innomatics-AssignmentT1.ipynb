## Problem 1: User Login Check

```python
# Step 1: Store correct username and password
correct_username = "admin"
correct_password = "1234"

# Step 2: Take login input from the user
username = input("Enter username: ")
password = input("Enter password: ")

# Step 3: Check whether both username and password match
if username == correct_username and password == correct_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

## Problem 2: Pass / Fail Analyzer

```python
# Step 1: List of student marks
marks = [45, 78, 90, 33, 60]

# Step 2: Initialize counters
pass_count = 0
fail_count = 0

# Step 3: Check each mark
for m in marks:
    
    # Student passes if marks >= 50
    if m >= 50:
        pass_count = pass_count + 1
    else:
        fail_count = fail_count + 1

# Step 4: Display the result
print("Total Students:", len(marks))
print("Pass Students:", pass_count)
print("Fail Students:", fail_count)
```

## Problem 3: Simple Data Cleaner

```python
# Step 1: Original list of names
names = [" Alice ", "bob", " CHARLIE "]

# Step 2: Create an empty list to store cleaned names
cleaned_names = []

# Step 3: Loop through each name and clean it
for name in names:
    
    # Remove extra spaces using strip()
    name = name.strip()
    
    # Convert to lowercase
    name = name.lower()
    
    # Add cleaned name to new list
    cleaned_names.append(name)

# Step 4: Display cleaned list
print("Cleaned Names:", cleaned_names)
```

## Problem 4: Message Length Analyzere for clear output


```python
# Step 1: List of messages
messages = ["Hi", "Welcome to the platform", "OK"]

# Step 2: Loop through each message
for msg in messages:
    
    # Find length of the message
    length = len(msg)
    
    # Print the message and its length
    print("Message:", msg)
    print("Length:", length)
    
    # Flag messages longer than 10 characters
    if length > 10:
        print("Status: Message is longer than 10 characters")
    
    print()   # Empty line for clear output
```

## Problem 5: Error Message Detector

```python
# Step 1: List of system logs
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Step 2: Initialize error counter
error_count = 0

# Step 3: Loop through logs to detect "ERROR"
for log in logs:
    
    # Check if current log is an error
    if log == "ERROR":
        error_count = error_count + 1

# Step 4: Display total error count
print("Total ERROR entries:", error_count)
```
