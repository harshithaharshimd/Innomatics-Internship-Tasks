## Problem 1: Unique Words in a Sentence


```python
# Taking input from the user
sentence = input("Enter sentence: ")

# Step 1: Convert the sentence into lowercase 
sentence = sentence.lower()

# Step 2: Split the sentence into individual words
words_list = sentence.split()

# Step 3: Convert list into a set to remove duplicate words
unique_words = set(words_list)

# Step 4: Find the count of unique words
unique_count = len(unique_words)

# Step 5: Display the result
print("Unique words count:", unique_count)
print("Unique words:", unique_words)
```

## Problem 2: Highest Salary from Employee Data

```python
# Step 1: Store employee names and their salaries in a dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Step 2: Assume the first employee has the highest salary initially
highest_employee = None
highest_salary = 0

# Step 3: Traverse through the dictionary to compare salaries
for name in employees:
    salary = employees[name]   
    
    # Check if current salary is greater than the stored highest salary
    if salary > highest_salary:
        highest_salary = salary
        highest_employee = name

# Step 4: Display the result
print("Highest Salary:", highest_employee, "-", highest_salary)
```

## Problem 3: Find Maximum and Minimum Values


```python
# Step 1: Define the list of numbers
numbers = [45, 22, 89, 10, 66]

# Step 2: Initialize max and min with the first element
maximum_value = numbers[0]
minimum_value = numbers[0]

# Step 3: Loop through the list to compare values
for num in numbers:
    
    # Check for maximum value
    if num > maximum_value:
        maximum_value = num
    
    # Check for minimum value
    if num < minimum_value:
        minimum_value = num

# Step 4: Print the results
print("List:", numbers)
print("Max:", maximum_value)
print("Min:", minimum_value)
```

## Problem 4: Count Products Above a Price Threshold

```python
# Step 1: Define the list of product prices
prices = [450, 1200, 899, 1500, 300]

# Step 2: Initialize a counter
count = 0

# Step 3: Loop through each price and check if it is greater than 1000
for price in prices:
    
    # If the condition is true, increase the counter
    if price > 1000:
        count = count + 1

# Step 4: Display the result
print("Products above 1000:", count)
```

## Problem 5: Calculate Attendance Percentage

```python
# Step 1: Define the attendance list
attendance = ["P", "P", "A", "P", "P"]

# Step 2: Initialize counter for present days
present_count = 0

# Step 3: Count how many times "P" appears
for status in attendance:
    
    # Check if the student/employee was present
    if status == "P":
        present_count = present_count + 1

# Step 4: Calculate total days
total_days = len(attendance)

# Step 5: Calculate attendance percentage
attendance_percentage = (present_count / total_days) * 100

# Step 6: Display result
print("Attendance Percentage:", attendance_percentage)
```

## Problem 7: Count Character Frequency


```python
# Step 1: Take input string from user
text = input("Enter a string: ")

# Step 2: Create an empty dictionary to store character counts
char_frequency = {}

# Step 3: Loop through each character in the string
for ch in text:
    
    # If character already exists in dictionary, increase count
    if ch in char_frequency:
        char_frequency[ch] = char_frequency[ch] + 1
    else:
        # Otherwise add new character with count 1
        char_frequency[ch] = 1

# Step 4: Display the result
print(char_frequency)
```

## Problem 10: Calculate Average Marks

```python
# Step 1: Define the list of marks
marks = [65, 75, 80, 70]   

# Step 2: Initialize total sum
total = 0

# Step 3: Add all marks using a loop
for m in marks:
    total = total + m

# Step 4: Calculate average
average_marks = total / len(marks)

# Step 5: Display result
print("Average Marks:", average_marks)
```
